```python
def getDummies(df):
    catCol = ["MSSubClass", "MSZoning", "Street", "Alley", "LotShape", "LandContour", "Utilities", "LotConfig", "LandSlope", "Neighborhood", "Condition1", "Condition2", "BldgType", "HouseStyle", "RoofStyle", "RoofMatl", "Exterior1st", "Exterior2nd", "MasVnrType", "ExterQual", "ExterCond", "Foundation", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "Heating", "HeatingQC", "CentralAir", "Electrical", "KitchenQual", "Functional", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "PavedDrive", "PoolQC", "Fence", "SaleType", "SaleCondition"]
    df = pd.get_dummies(df, columns=catCol)
    return df
```

```python
# FILL MISSING VALUES
def fillMissing(df):
    # Missing Values in train set
    
    # LotFrontage - Fill with mean
    df["LotFrontage"] = df["LotFrontage"].fillna(df["LotFrontage"].mean())
    # Alley - Fill "NA" with "None"
    df["Alley"] = df["Alley"].fillna("None")
    # MasVnrType - Fill with mode
    df["MasVnrType"] = df["MasVnrType"].fillna(df["MasVnrType"].mode().iloc[0])
    # MasVnrArea - Fill with mean
    df["MasVnrArea"] = df["MasVnrArea"].fillna(df["MasVnrArea"].mean())
    # BsmtQual - Fill with "None"
    df["BsmtQual"] = df["BsmtQual"].fillna("None")
    # BsmtCond - Fill with "None"
    df["BsmtCond"] = df["BsmtCond"].fillna("None")
    # BsmtExposure - Fill with "None"
    df["BsmtExposure"] = df["BsmtExposure"].fillna("None")
    # BsmtFinType1 - Fill with "None"
    df["BsmtFinType1"] = df["BsmtFinType1"].fillna("None")
    # BsmtFinType2 - Fill with "None"
    df["BsmtFinType2"] = df["BsmtFinType2"].fillna("None")
    # Electrical - Fill with mode
    df["Electrical"] = df["Electrical"].fillna(df["Electrical"].mode().iloc[0])
    # GarageType - Fill with "None"
    df["GarageType"] = df["GarageType"].fillna("None")
    # GarageYrBlt - Fill with same value as "YearBuilt"
    df["GarageYrBlt"] = df["GarageYrBlt"].fillna(df["YearBuilt"])
    # GarageFinish - Fill with "None"
    df["GarageFinish"] = df["GarageFinish"].fillna("None")
    # GarageQual - Fill with "None"
    df["GarageQual"] = df["GarageQual"].fillna("None")
    # GarageCond - Fill with "None"
    df["GarageCond"] = df["GarageCond"].fillna("None")
    # PoolQC - Fill with "None"
    df["PoolQC"] = df["PoolQC"].fillna("None")
    # Fence -  Fill with "None"
    df["Fence"] = df["Fence"].fillna("None")
    # MiscFeature - Fill with "None"
    df["MiscFeature"] = df["MiscFeature"].fillna("None")
    
    # Missing Values in Test Set
    
    # Utilities - Fill with Mode
    df["Utilities"] = df["Utilities"].fillna(df["Utilities"].mode().iloc[0])
    # Exterior1st - Fill with Mode
    df["Exterior1st"] = df["Exterior1st"].fillna(df["Exterior1st"].mode().iloc[0])
    # Exterior2nd - Fill with Mode
    df["Exterior2nd"] = df["Exterior2nd"].fillna(df["Exterior2nd"].mode().iloc[0])
    # BsmtFinSF1 - Fill with mean
    df["BsmtFinSF1"] = df["BsmtFinSF1"].fillna(df["BsmtFinSF1"].mean())
    # BsmtFinSF2 - Fill with mean
    df["BsmtFinSF2"] = df["BsmtFinSF2"].fillna(df["BsmtFinSF2"].mean())
    # BsmtUnfSF - Fill with mean
    df["BsmtUnfSF"] = df["BsmtUnfSF"].fillna(df["BsmtUnfSF"].mean())
    # TotalBsmtSF - Fill with mean
    df["TotalBsmtSF"] = df["TotalBsmtSF"].fillna(df["TotalBsmtSF"].mean())
    # BsmtFullBath - Fill with mode
    df["BsmtFullBath"] = df["BsmtFullBath"].fillna(df["BsmtFullBath"].mode().iloc[0])
    # BsmtHalfBath - Fill with mode
    df["BsmtHalfBath"] = df["BsmtHalfBath"].fillna(df["BsmtHalfBath"].mode().iloc[0])
    # KitchenQual - Fill with mode
    df["KitchenQual"] = df["KitchenQual"].fillna(df["KitchenQual"].mode().iloc[0])
    # Functional - Fill with mode
    df["Functional"] = df["Functional"].fillna(df["Functional"].mode().iloc[0])
    # GarageCars - Fill with mode
    df["GarageCars"] = df["GarageCars"].fillna(df["GarageCars"].mode().iloc[0])
    # GarageArea - Fill with mean
    df["GarageArea"] = df["GarageArea"].fillna(df["GarageArea"].mean())
    # SaleType - Fill with mode
    df["SaleType"] = df["SaleType"].fillna(df["SaleType"].mode().iloc[0])
    return df
```

```python
# REMOVE UNNECESSARY FEATURES
def removeUnnecessary(df):
    df.pop("Id")
    df.pop("FireplaceQu") # Fireplace not really important
    df.pop("MiscFeature") # MiscVal makes this redundant
    df.pop("MoSold")      # Month sold not really important
```

```python
def standardize(df):
    scaler = StandardScaler().fit(df)
    standardized_x = scaler.transform(df)
    return standardized_x
```

```python
# PREPROCESS
combined = pd.concat((x, x_test), axis=0)
combined = fillMissing(combined)
removeUnnecessary(combined)
combined = getDummies(combined)
combined = pd.DataFrame(standardize(combined))

# SPLIT INTO X, X_TEST
x = combined.iloc[:1460, :]
x_test = combined.iloc[1460:, :].values
```

```python
# SPLIT INTO DATA FOR TRAINING
x_train, x_val, y_train, y_val = train_test_split(x.values, y.values, test_size=0.2)
```

```python
# CREATE MODEL - 2 Layer NN
model = Sequential()
model.add(Dense(256, activation='relu', input_shape=(305,)))
model.add(Dense(512, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(2048, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mean_squared_logarithmic_error', metrics=['mean_squared_error'])
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=32)
```

```python
model.evaluate(x_val, y_val)
```

```python
def saveModel(model):
    # Column 1: "ImageId"
    id_ = [(i+1) for i in range(1460, 2919)]

    # Column 2: "Label"
    predictions = model.predict(x_test)

    # Save
    np.savetxt("sub_4.csv", X=np.c_[id_, predictions], fmt=['%i', '%f'], delimiter=',', header="Id,SalePrice", comments='')
```

```python
# Predict and Save as CSV
saveModel(model)
# Save Model
model.save('model1.h5')
```
