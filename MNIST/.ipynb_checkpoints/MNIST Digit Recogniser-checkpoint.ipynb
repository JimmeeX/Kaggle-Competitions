```python
%matplotlib inline
# IMPORT PYTHON LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

First, let's have a look to see what data we are playing with. I will load up the data (downlaoded from Kaggle) as a dataframe using pandas *pd.read_csv()* function.

```python
df = pd.read_csv("train.csv")
display(df.info())
display(df.describe())
display(df)
```

The table above indicates that we have 42000 rows, and 785 columns. The first column is *label*, and the values range from 0-9.

```python
labelVals = list(df["label"].unique())
labelVals.sort()
display(labelVals)
```

The other columns (*pixel 0 to pixel 783*) each represents a single pixel from a 28x28 picture. The values range from (0 ~ 255). Let's have a look at what the first 100 rows depict.

```python
# To separate df into y (labels) and x (variables/features)
y = df.pop("label")
x = df / 255 #Normalize to [-1, 1] (assuming 0 is min and 255 is max)
```

```python
def displayDigits(x, n):
    fig = plt.figure()
    for i in range(n):
        # Reshape Matrix from 784x1 to 28x28
        matrix = x.loc[i, :].values.reshape(28,28)
        # Add image to subplot
        plotFig = fig.add_subplot(10, 10, i+1)
        plotFig.axis('off')
        plt.imshow(matrix, cmap='gray')
displayDigits(df, 100)
```

Here, we can see that each image represents a handwritten digit, ranging from 0 ~ 9.

The first step to training a neural network is to decide the architecture of the neural network. 

The **input layer** will have **784 nodes**, where each node represents each pixel. 

The **output layer** will have **10 nodes**, where each node represents the number of possible classifications (0,1,2,3...8,9).

For this exercise, I will include a **single hidden layer** with **20 nodes**, but this can be changed and experimented with.

```python
# Get Parameters
INPUT_LAYER = x.shape[1]
OUTPUT_LAYER = 10
HIDDEN_LAYER = 20
IMAGE_SIZE = 28

temp = pd.DataFrame(np.eye(OUTPUT_LAYER))
target = temp.loc[y, :]

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2)
```

```python
# Initialise Weights
# First matrix of weights represents all the possible paths from the INPUT_LAYER to the HIDDEN_LAYER, and should be a
# matrix of size 20x784.
# Weights are initialised under normal distribution
np.random.seed(seed=13)
Theta_1 = np.random.normal(size=(HIDDEN_LAYER, INPUT_LAYER+1)) # 20 x 784+1
Theta_2 = np.random.normal(size=(OUTPUT_LAYER, HIDDEN_LAYER+1)) # 10 x 20+1
```

```python
# Define our sigmoid activation function
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
# Define our sigmoid gradient function
def sigmoidGradient(x):
    temp = np.exp(-x)
    return np.divide(temp, (1+temp)**2)
```

```python
def relu(x):
    return (x * (x > 0))
```

```python
def reluGradient(x):
    return (x>0).astype("double")
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def tanhGradient(x):
    return (1-(np.tanh(x))**2)
```

```python
# Forward Propagate
# Calculate Cost Function
# Back Propagate
def twoLayer(Theta_1, Theta_2, input_layer_size, hidden_layer_size, output_layer_size, x, y, lambda_, activ, activGrd):
    m = x.shape[0]   # 33600
    m_ones = np.ones((m, 1)) ## 33600 x 1
    
    # Forward Propagate
    a1 = np.concatenate((m_ones, x.values), axis=1)
    z2 = np.dot(a1, Theta_1.transpose())
    a2 = np.concatenate((m_ones, activ(z2)), axis=1)
    z3 = np.dot(a2, Theta_2.transpose())
    a3 = activ(z3)
    
    # Create Matrix for Target Values
    eyeMatrix = pd.DataFrame(np.eye(output_layer_size))
    target = eyeMatrix.loc[y, :]
    
    # Calculate Cost (without regularisation)
    cost = 0
    #temp1 = np.multiply(target, np.log(a3))
    #temp2 = np.multiply(1-target, np.log(1-a3))
    #temp3 = np.sum(np.sum(Theta_1[:, 1:]**2))
    #temp4 = np.sum(np.sum(Theta_2[:, 1:]**2))
    #cost = - 1 / m * np.sum(np.sum(temp1 + temp2)) + lambda_ / 2 / m * (temp3+temp4)
    
    # Back Propagate (without regularisation)
    d3 = a3-target
    temp1 = np.dot(d3, Theta_2[:, 1:])
    d2 = np.multiply(temp1, activGrd(z2))
    delta1 = np.dot(d2.transpose(), a1) # 20 x 784
    delta2 = np.dot(d3.transpose(), a2) # 10 x 20
    Theta1_grad = 1 / m * delta1
    Theta2_grad = 1 / m * delta2
    Theta_1[:, 0] = 0
    Theta_2[:, 0] = 0
    Theta_1 = Theta_1 * lambda_ / m
    Theta_2 = Theta_2 * lambda_ / m
    Theta1_grad = Theta1_grad + Theta_1
    Theta2_grad = Theta2_grad + Theta_2
    
    return (cost, Theta1_grad, Theta2_grad)
```

```python
def trainTwoLayer(Theta_1, Theta_2, Theta1_grad, Theta2_grad, learning_rate):
    Theta_1 = Theta_1 - learning_rate * Theta1_grad
    Theta_2 = Theta_2 - learning_rate * Theta2_grad
    return (Theta_1, Theta_2)
```

```python
def predictTwoLayer(Theta_1, Theta_2, x, activ):
    m = x.shape[0]
    m_ones = np.ones((m, 1))
    
    # Forward Propagate
    a1 = np.concatenate((m_ones, x.values), axis=1)
    z2 = np.dot(a1, Theta_1.transpose())
    a2 = np.concatenate((m_ones, tanh(z2)), axis=1)
    z3 = np.dot(a2, Theta_2.transpose())
    a3 = activ(z3)
    pred = a3.argmax(axis=1)
    return pred
```

```python
def accuracy(x, y, m):
    return (np.sum(x==y) / m)
```

```python
#def trainTwoLayer(Theta_1, Theta_2, input_layer_size, hidden_layer_size, output_layer_size, x_train, y_train, lambda_, learning_rate)
costArray = []
accArray = []
learning_rate = 0.5
lambda_ = 1
for numIterations in range(50):
    # FORWARD AND BACK PROPAGATE
    cost, Theta1_grad, Theta2_grad = twoLayer(Theta_1, Theta_2, INPUT_LAYER, HIDDEN_LAYER, OUTPUT_LAYER, x_train, y_train, lambda_, sigmoid, sigmoidGradient)
    # UPDATE THETA
    Theta_1, Theta_2 = trainTwoLayer(Theta_1, Theta_2, Theta1_grad, Theta2_grad, learning_rate)
    # GET PREDICTIONS
    pred = predictTwoLayer(Theta_1, Theta_2, x_val, sigmoid)
    # GET ACCURACY
    acc = accuracy(pred, y_val, x_val.shape[0])
    costArray.append(cost)
    accArray.append(acc)
    # display(cost)
    display(acc)
```
